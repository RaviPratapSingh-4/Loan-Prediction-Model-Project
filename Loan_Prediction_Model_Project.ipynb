```python
import pandas as pd
```

# Reading Dataset

```python
df = pd.read_csv(r"C:\Users\RAVI\Videos\Downloads\loan _data_after_feature_enginnering.csv")
```

```python
df
```

```python
df.info()
```

```python
columns_to_extract = [
    "ApplicantIncome", 
    "CoapplicantIncome", 
    "LoanAmount", 
    "Loan_Amount_Term", 
    "Credit_History", 
    "Total_Income", 
    "LTI"
]
```

```python
x = df[columns_to_extract]
```

```python
x
```

```python
y = df["Loan_Status"]
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size = 0.30, random_state = 4)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
var = StandardScaler()
x_train = var.fit_transform(x_train)
x_test = var.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
logReg = LogisticRegression()
logReg.fit(x_train, y_train)
y_pred = logReg.predict(x_test)
```

```python
y_pred
```

# Results:

# Accuracy Score of the Logistic Regression model

```python
accuracy_score(y_pred, y_test)
```

# Confusion Matrix of the Logistic Regression model

```python
confusion_matrix(y_pred, y_test)
```

# Brief on the numerical result 

# the results shows that applicants who have more probability of getting loan have :
# i- Applicants with higher ApplicantIncome and CoapplicantIncome 
# ii- Individuals with lower Loan-to-Income (LTI) ratios 
# iii- applicants with Good Credit History
# iv- Applicants requesting a LoanAmount that is reasonable relative to their income and loan term
# v- Loans with longer repayment periods (Loan_Amount_Term) might be more favorable as they reduce the monthly repayment burden, increasing affordability
